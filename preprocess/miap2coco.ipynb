```python
import pandas as pd
import os
import PIL.Image as Image
```

```python
df = pd.read_csv('/home/prml/Dataset/MIAP/annotations/open_images_extended_miap_boxes_val.csv')
df.head()

data_type = 'val'
df.head()
```

```python
df['ImageID'].values
```

```python
images = {}
image_id = 1

for ImageID in df['ImageID'].values:
    if ImageID not in images:
        width, height = Image.open(os.path.join(f'/home/prml/Dataset/MIAP/{data_type}', ImageID + '.jpg')).size
        images[ImageID] = {'file_name': ImageID + '.jpg', 'height':height, 'width': width, 'id': image_id}
        image_id += 1

len(images)
```

```python
annotations = []
ann_id = 1
for i, row in df.iterrows():
    ImageID = row['ImageID']
    width, height = images[ImageID]['width'], images[ImageID]['height']
    x, y, w, h = int(row['XMin'] * width), int(row['YMin'] * height), int((row['XMax'] - row['XMin']) * width), int((row['YMax'] - row['YMin']) * height)
    print(x, y, w, h)
    annotations.append({'image_id': images[ImageID]['id'], 'bbox': [x, y, w, h], 'category_id': 1, 'id': ann_id, 'iscrowd': 0, 'area':w*h})
    ann_id += 1

len(annotations)
```

```python
import json


# Create COCO format dictionary
converted = {
    'info': {"description":'MIAP', 'url':'https://storage.googleapis.com/openimages/web/extended.html#miap', 'version':'1.0', 'year':2021, 'contributor':'MIAP', 'date_created':'2021-06-01'},
    'licenses': [],
    'images': images,
    'annotations': annotations,
    'categories': [{'supercategory':'person', 'id':1, 'name':'person'}]
}

# Convert COCO format dictionary to JSON
coco_json = json.dumps(converted)

# Save JSON to a file
with open('annotations.json', 'w') as f:
    f.write(coco_json)
```
